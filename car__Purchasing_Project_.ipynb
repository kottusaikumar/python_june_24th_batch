```python
import numpy as np
import pandas as pd
import seaborn
import sys
import math
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


class TRAINING:
    def __init__(self, location):
        try:
            self.df = pd.read_csv(location, encoding='ISO-8859-1')
            self.df = self.df.drop(["Customer Name", "Customer e-mail", "Country"], axis=1)
            self.X = self.df.iloc[:, :-1]
            self.y = self.df.iloc[:, -1]
            self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, train_size=0.2, random_state=32)
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def training_data(self):
        try:
            self.reg = LinearRegression()
            self.reg.fit(self.X_train, self.y_train)
            self.Training_data = pd.DataFrame(self.X_train.copy())
            self.Training_data["Actual_data"] = self.y_train
            self.Training_data["Prediction_data"] = self.reg.predict(self.X_train)
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def mean_squared_error_train(self):
        try:
            self.count = 0
            for i in self.Training_data.index:
                self.count += (self.Training_data["Actual_data"][i] - self.Training_data["Prediction_data"][i])**2
            print(f"MSE_accuracy_train = {self.count / len(self.Training_data)}")
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def r2_score_train(self):
        try:
            self.count = 0
            self.num = 0
            self.mean_data = 0
            for j in self.y_train:
                self.mean_data += j
            self.mean_data = self.mean_data / len(self.y_train)
            for i in self.Training_data.index:
                self.count += (self.Training_data["Actual_data"][i] - self.Training_data["Prediction_data"][i])**2
                self.num += (self.Training_data["Actual_data"][i] - self.mean_data)**2
            r2_score = 1 - (self.count / self.num)
            print(f"r2_score_accuracy_train = {r2_score}")
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")


    def root_mean_squared_error_train(self):
        try:
            self.count = 0
            for i in self.Training_data.index:
                self.count += (self.Training_data["Actual_data"][i] - self.Training_data["Prediction_data"][i])**2
            print(f"RMSE_accuracy_train = {math.sqrt(self.count / len(self.Training_data))}")
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def training_performance(self):
        try:

           self.r2_score_train()
           self.mean_squared_error_train()
           self.root_mean_squared_error_train()
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def testing_data(self):
        try:
            self.Testing_data = pd.DataFrame(self.X_test.copy())
            self.Testing_data["Actual_data"] = self.y_test
            self.Testing_data["Prediction_data"] = self.reg.predict(self.X_test)
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def r2_score_test(self):
        try:
            self.count = 0
            self.num = 0
            self.mean_data = 0
            for j in self.y_test:
                self.mean_data += j
            self.mean_data = self.mean_data / len(self.y_test)
            for i in self.Testing_data.index:
                self.count += (self.Testing_data["Actual_data"][i] - self.Testing_data["Prediction_data"][i])**2
                self.num += (self.Testing_data["Actual_data"][i] - self.mean_data)**2
            r2_score = 1 - (self.count / self.num)
            print(f"r2_score_accuracy_test = {r2_score}")
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def mean_squared_error_test(self):
        try:
            self.count = 0
            for i in self.Testing_data.index:
                self.count += (self.Testing_data["Actual_data"][i] - self.Testing_data["Prediction_data"][i])**2
            print(f"MSE_accuracy_test = {self.count / len(self.Testing_data)}")
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")


    def root_mean_squared_error_test(self):
        try:
            self.count = 0
            for i in self.Testing_data.index:
                self.count += (self.Testing_data["Actual_data"][i] - self.Testing_data["Prediction_data"][i])**2
            print(f"RMSE_accuracy_test = {math.sqrt(self.count / len(self.Testing_data))}")
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")

    def test_performance(self):
        try:
           self.r2_score_test()
           self.mean_squared_error_test()
           self.root_mean_squared_error_test()
        except Exception as e:
            error_type, error_msg, error_line = sys.exc_info()
            print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")


if __name__ == "__main__":
    try:
        obj = TRAINING("Car_Purchasing_Data.csv")
        print(f"training_accuracy and training_loss_data:")
        print("")
        obj.training_data()

        obj.training_performance()
        print("")
        print(f"testing_accuracy and testing_loss_data:")
        print("")
        obj.testing_data()

        obj.test_performance()
    except Exception as e:
        error_type, error_msg, error_line = sys.exc_info()
        print(f"Error from line {error_line.tb_lineno} -> type {error_type} -> msg {error_msg}")
```
